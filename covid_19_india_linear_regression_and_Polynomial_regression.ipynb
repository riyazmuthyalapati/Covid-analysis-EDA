```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Covid19india.csv")
df.head()
```

```python
df = df.iloc[:,1:]
df.head()
```

```python
Day = df[df["Current Status"] == "Hospitalized"].groupby(["Month","Day"])[["Num Cases"]].sum()
Day
```

```python
x = np.arange(len(Day))
x
```

```python
y = Day.values
y
```

```python
plt.scatter(x,y)
```

```python
#reshaping x variable 
x
```

```python
x = x.reshape(-1,1)
x
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x,y)
```

```python
y[121]
```

```python
regressor.predict([[122]])
```

```python
regressor.score(x,y)
```

```python
regressor.predict([[121]])
```

```python
y_predicted = regressor.predict(x)
y_predicted
```

```python
plt.scatter(x,y)
plt.plot(x,y_predicted)
```

## Predecting using Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=5)
X = poly.fit_transform(x.reshape(-1,1))
X
```

```python
pd.DataFrame(X)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
YP_P = reg.predict(X)
YP_P
```

```python
plt.scatter(x,y)
plt.plot(x,YP_P, color = "k")
```

```python
reg.score(X,y)*100
```
